```python
# 分组计算，group by
 
import pandas
import pandas as pd 
df = pd.read_csv('E:\movie.csv')
df.head(10)
df.shape

# 统计，每个导演有多少电影
df.director_name.value_counts()

# 分组，按导演名字分组
grouped = df.groupby('director_name')
grouped # 是一个对象，有多个属性和方法
```

```python
# 分组后的组信息
grouped.size() #有多少组，每个组中有多少条数据

grouped.groups #组的数据结构是什么

len(grouped) # 有多少组
```

```python
for name,group in grouped:
    print(name)
    print(type(group))
```

```python
# 分组的统计计算
# 分组计算每一个统计函数都是作用在每个group上

grouped.mean() #每个数值型数据在每个组中的平均值

grouped.sum()

grouped.std()
```

```python
# 只针对某个或某几个特征进行计算
grouped['duration','director_facebook_likes'].sum()
```

```python
# 对每一个数据特征都求 平均值，总和，标准差
import numpy as np
grouped.agg([np.mean, np.sum, np.std])
grouped['duration','director_facebook_likes'].agg([np.mean, np.sum, np.std])
```

```python
# 不同的列用不同的统计量
grouped.agg({'duration':np.mean,'actor_1_facebook_likes':np.sum})
```

```python
# transformation 对组数据做标准化或填充缺省值

# 对组数值型数据做标准化处理
# （数值-平均数）/ 标准差

# 由于原数值中有空值，会造成数据的不一致，因此用fillna(0)填充0值
df2 = df.fillna(0)
grouped = df2.groupby('director_name')

grouped.filter(grouped['duration'] grouped['duration'] > 150 )



#.filter(lambda g: g.mean() > 150)
# filtration 对组数据做过滤处理
```

```python
# 定义一个标准化函数
z_score = lambda s : (s - s.mean()) / s.std()
```

```python
df2.shape
```

```python
grouped['duration','actor_1_facebook_likes'].transform(z_score)

grouped['duration'].filter(lambda g: g.mean() > 150)
```

```python
# 数据过滤

# 每个组里面小于2的剔除
grouped.filter(lambda g: len(g) > 1)

grouped.filter(lambda g: len(g) > 1)['director_name'].value_counts()
```
